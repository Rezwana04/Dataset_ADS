# Ego_Vehicle_Ops_Value column contains the information related to Ego vehicle(e.g., spped, accelaration)

```python
df_ego = data[['test_scenario_number','Ego_Vehicle_Ops_Value']]
df_ego
```

# spliting the string to extract necessary information

```python
import re
new_rows_list  = []

###-----------iterate the rows for each commit
for index, row in df_ego.iterrows():
    stat=row['Ego_Vehicle_Ops_Value']
    stat_array = stat.split(',')
    
    ###-----------search for acceleration, throttle, steering_target, brake, steering_rate and speed text to store into dataframe columns
    #acceleration
    acceleration = [s for s in stat_array if 'acceleration' in s]
    if not acceleration:
        acceleration = 0
    else:
        acceleration = acceleration[0]
    #throttle
    throttle= [s for s in stat_array if 'throttle' in s]
    if not throttle:
        throttle = 0
    else:
        throttle = throttle[0]
        
    #steering_target
    steering_target= [s for s in stat_array if 'steering_target' in s]
    if not steering_target:
        steering_target = 0
    else:
        steering_target = steering_target[0]
    #brake     
    brake = [s for s in stat_array if 'brake' in s]
    if not brake:
        brake = 0
    else:
        brake = brake[0]
    #steering_rate
    steering_rate= [s for s in stat_array if 'steering_rate' in s]
    if not steering_rate:
        steering_rate = 0
    else:
        steering_rate = steering_rate[0]
        
    #speed
    speed= [s for s in stat_array if 'speed' in s]
    if not speed:
        speed = 0
    else:
        speed = speed[0]
    
        
    new_rows_list.append([row['Ego_Vehicle_Ops_Value'],acceleration, throttle, steering_target, brake, steering_rate, speed])
    
```

```python
df = pd.DataFrame (new_rows_list, columns=['Ego_Vehicle_Ops_Value', 'acceleration', 'throttle', 'steering_target', 'brake', 'steering_rate', 'speed'])
df.head()
```

# Taking only the float value

```python
df['acceleration'] = df['acceleration'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
df['throttle'] = df['throttle'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
df['steering_target'] = df['steering_target'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
df['brake'] = df['brake'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
df['steering_rate'] = df['steering_rate'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
df['speed'] = df['speed'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
```

```python
df_ego=df
df_ego
```

# Scenario column contains the information related to the environment(e.g., weather, number of pedestrian, number of other vehicle)

```python
df_scenario = data[['test_scenario_number','Scenario']]
df_scenario
```

```python
import re
new_rows_list  = []

###-----------iterate the rows for each commit
for index, row in df_scenario.iterrows():
    stat=row['Scenario']
    stat_array = stat.split(',')
    
    ###-----------search for Ego_Operation, Ego_Speed, TimeofDay,TrafficRule_Sidewalk and CollisionInfoAtTimeStep text to store into dataframe columns
   
    #Ego_Operation
    Ego_Operation = [s for s in stat_array if 'Ego_Operation' in s]
    if not Ego_Operation:
        Ego_Operation = 0
    else:
        Ego_Operation = Ego_Operation[0]
        
    #Ego_Speed
    Ego_Speed= [s for s in stat_array if 'Ego_Speed' in s]
    if not Ego_Speed:
        Ego_Speed = 0
    else:
        Ego_Speed = Ego_Speed[0]
        
    #TimeofDay
    TimeofDay= [s for s in stat_array if 'TimeofDay' in s]
    if not TimeofDay:
        TimeofDay = 0
    else:
        TimeofDay = TimeofDay[0]
    
    #CollisionInfoAtTimeStep
    CollisionInfoAtTimeStep= [s for s in stat_array if 'CollisionInfoAtTimeStep' in s]
    if not CollisionInfoAtTimeStep:
        CollisionInfoAtTimeStep = 0
    else:
        CollisionInfoAtTimeStep = CollisionInfoAtTimeStep[0]
        
     
    #TrafficRule[Sidewalk]
    TrafficRule_Sidewalk= [s for s in stat_array if 'TrafficRule[Sidewalk]' in s]
    if not TrafficRule_Sidewalk:
        TrafficRule_Sidewalk = 0
    else:
        TrafficRule_Sidewalk = TrafficRule_Sidewalk[0]
        
    
    
        
    new_rows_list.append([row['Scenario'],Ego_Operation,Ego_Speed, TimeofDay,TrafficRule_Sidewalk,CollisionInfoAtTimeStep])
    
```

```python
df = pd.DataFrame (new_rows_list, columns=['scenario','Ego_Operation','Ego_Speed', 'TimeofDay','TrafficRule_Sidewalk','CollisionInfoAtTimeStep'])
df.head()
```

```python
df = df[['Ego_Operation','Ego_Speed','TimeofDay','TrafficRule_Sidewalk','CollisionInfoAtTimeStep']].copy()
df
```

```python
df['ego_speed'] = df.Ego_Speed.str.extract(':(.*)')
df['timeofDay'] = df.TimeofDay.str.extract(':(.*)')
df['trafficRule_sidewalk'] = df.TrafficRule_Sidewalk.str.extract(':(.*)')
df['CollisionInfo'] = df.CollisionInfoAtTimeStep.str.extract(':(.*)')
df['ego_operation'] = df.Ego_Operation.str.extract(':(.*)')
df
```

```python
first_words = df.apply(lambda x: x.str.split().str[0])
first_words
```

```python
first_words=first_words.replace({'\'': ' '}, regex=True)
first_words
```

```python
df_collision = first_words[['ego_operation','ego_speed','timeofDay','trafficRule_sidewalk','CollisionInfo']].copy()
df_collision
```

# weather Preprocessing

```python
import re
new_rows_list  = []

###-----------iterate the rows for each commit
for index, row in df_scenario.iterrows():
    stat=row['Scenario']
    stat_array = stat.split(',')
    
    ###-----------search for Ego_Operation, Ego_Speed, Pedestrian, NPC, Static_obstacle, TimeofDay,TrafficRule_Sidewalk and CollisionInfoAtTimeStep text to store into dataframe columns
   
    #Weather_rain
    Weather_rain = [s for s in stat_array if 'Weather[rain]' in s]
    if not Weather_rain:
        Weather_rain = 0
    else:
        Weather_rain = Weather_rain[0]
        
    #Weather_fog
    Weather_fog= [s for s in stat_array if 'Weather[fog]' in s]
    if not Weather_fog:
        Weather_fog = 0
    else:
        Weather_fog = Weather_fog[0]
        
    #Weather_wetness
    Weather_wetness= [s for s in stat_array if 'Weather[wetness]' in s]
    if not Weather_wetness:
        Weather_wetness = 0
    else:
        Weather_wetness = Weather_wetness[0]
        
    new_rows_list.append([row['Scenario'],Weather_rain,Weather_fog, Weather_wetness])
    
```

```python
weather = pd.DataFrame (new_rows_list, columns=['scenario','Weather_rain','Weather_fog', 'Weather_wetness'])
weather.head()
```

```python
df_weather=weather[['Weather_rain','Weather_fog','Weather_wetness']]
df_weather=df_weather.apply(lambda x: x.str.split().str[1])
df_weather=df_weather.replace({'\'': ' '}, regex=True)
df_weather
```

# pedestrian Preprocessing

```python
pedestrian=df_scenario[["Scenario"]]
words = ["Pedestrian"]

pedestrian["pedestrian_count"] = pedestrian.Scenario.apply(lambda x: sum([x.count(word) for word in words]))
pedestrian
```

# vehicle Preprocessing

```python
vehicle=df_scenario[["Scenario"]]
words = ["NPC"]

vehicle["vehicle_count"] = vehicle.Scenario.apply(lambda x: sum([x.count(word) for word in words]))
vehicle
```

# combining all the dataframes

```python
result = pd.concat([df_ego,df_collision,df_weather,pedestrian,vehicle], axis=1)
result
```

```python
list = ['Ego_Vehicle_Ops_Value','Scenario','Scenario']
result=result.drop(list,axis = 1)
```

```python
result['pedestrian_count']=result['pedestrian_count'] - 1
result['vehicle_count']=result['vehicle_count'] - 1
```

# Mapping the values

```python
df_result=result
df_result['ego_speed'].replace({' None ':0,' Stop': 1,' Slow': 2,' Moderate':3,' Fast':4},inplace=True)
df_result['timeofDay'].replace({' Morning': 0,' Noon':1,' Night': 2},inplace=True)
df_result['trafficRule_sidewalk'].replace({' None ':0,' SlowDown ': 1},inplace=True)
df_result['CollisionInfo'].replace({' NotOccurred ':0,' BreakSafetyDistance ': 0,' Occurred ':1},inplace=True)
df_result['Weather_rain'].replace({' None':0 ,' Light':1,' Moderate':2, ' Heavy':3},inplace=True)
df_result['Weather_fog'].replace({' None':0, ' Light':1,' Moderate':2, ' Heavy':3,},inplace=True)
df_result['Weather_wetness'].replace({' None':0,' Light':1,' Moderate':2, ' Heavy':3,},inplace=True)
```

```python
df_result
```

```python
weather=df_result[['Weather_rain','Weather_fog','Weather_wetness']]
weather['Weather_rain'] = [0 if t <=0 else 1 for t in weather['Weather_rain']]
weather['Weather_fog'] = [0 if t <=0 else 2 for t in weather['Weather_fog']]
weather['Weather_wetness'] = [0 if t <=0 else 3 for t in weather['Weather_wetness']]
weather
```

```python
cols =['Weather_rain','Weather_fog','Weather_wetness']

weather ['weather'] = weather[cols].sum(axis=1)

weather
```

```python
weather.loc[weather['weather'] > 5] 
```

```python
weather=weather['weather']
df_result = pd.concat([df_result,weather], axis=1)
df_result
```

# Train_test split

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
X=df_result.drop(['CollisionInfo'],axis=1) # Features
y=df_result['CollisionInfo']
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

# Decission Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=123)
clf.fit(X_train,y_train)
y_pred_d=clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_d)
f1 = f1_score(y_test, y_pred_d)
recall = recall_score(y_test, y_pred_d)
precission = precision_score(y_test, y_pred_d)

print("Accuracy:", accuracy)
print("f1_score:", f1)
print("recall_score:", recall)
print("precision_score:", precission)
```

```python
confusion_matrix(y_test,y_pred_d)
print(pd.crosstab(y_test, y_pred_d, rownames = ['Actual'], colnames =['Predicted'], margins = True))
print(classification_report(y_test, y_pred_d))
```
